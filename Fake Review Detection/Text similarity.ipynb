```python
import numpy as np
import pandas as pd
import string

import nltk
from nltk.corpus import stopwords

from nltk.tokenize import word_tokenize
from nltk.stem.wordnet import WordNetLemmatizer
```

```python
doc1=" I had a factory Glock tool that I was using for my Glock 26, 27, and 17.  I've since lost it and had needed another.  Since I've used Ghost products prior, and know that they are reliable, I had decided to order this one.  Sure enough, this is just as good as a factory tool."

doc2=" I had a factory Glock tool that I was using for my Glock 26, 27, and 17.  I've since lost it and had needed another."
```

```python
stop_words= set(stopwords.words("english"))
print stop_words
```

```python
wnl=WordNetLemmatizer()
wnl.lemmatize("friendship")
```

```python
def filteration(text):
    word_tokens= word_tokenize(text)
    tokens= [w for w in word_tokens if w not in string.punctuation]
    filtered_sentence = [w for w in tokens if w not in stop_words]
    filt=[wnl.lemmatize(t) for t in filtered_sentence]
    return filt
```

```python
fil1=filteration(doc1)
```

```python
print fil1
```

```python
fil2= filteration(doc2)
```

```python
print fil2
```

```python
def similarity(text1, text2):
    uni= len(set(fil1).union(set(fil2)))
    intersect= len(set(fil1).intersection(set(fil2)))
    jaccard_similarity= (((1.0)*intersect)/uni)
    return jaccard_similarity

def checkDegree(number):
    if(number==1):
        return "identical"
    elif(number>=0.5):
        return "similar"
    elif(number==0):
        return "different"
    elif(number>0 and number<0.5):
        return "dissimilar"
```

```python
checkDegree(similarity(fil1, fil2))
```
