```python
from bs4 import BeautifulSoup
```

```python
import urllib2
```

```python
url="https://www.dineout.co.in/delhi/cyber-hub-social-dlf-cyber-city-gurgaon-22799/review?revpage=4"
```

```python
page=urllib2.urlopen(url)
```

```python
soup=BeautifulSoup(page)
```

```python
soup.prettify()
```

```python
all_rev=[]
all_rev= soup.find_all('span', attrs={'class':'more'})
```

```python
Review=[]
for rev in all_rev:
    #print rev.string
    Review.append(rev.string)
```

```python
all_names=[]
all_names= soup.find_all('div', attrs={'class': 'name'})
```

```python
Name=[]
for name in all_names:
    #print name.string
    Name.append(name.string)
```

```python
all_dates=[]
all_dates= soup.find_all('span', attrs={'class': 'date'})
```

```python
Date=[]
for date in all_dates:
    #print date.string[-11:]
    Date.append(date.string[-11:])
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(Name,columns=['Name'])
```

```python
df['Review']=Review
```

```python
df['Date']=Date
```
