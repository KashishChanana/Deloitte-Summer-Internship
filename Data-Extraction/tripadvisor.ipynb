```python
from bs4 import BeautifulSoup
```

```python
import urllib2
import pandas as pd
```

```python
url="https://www.tripadvisor.in/Restaurant_Review-g297615-d12175398-Reviews-Cyber_Hub_Social-Gurugram_Gurgaon_Gurgaon_District_Haryana.html"
```

```python
page=urllib2.urlopen(url)
```

```python
soup=BeautifulSoup(page)
```

```python
soup.prettify()
```

```python
all_rev=soup.find_all('p', attrs={'class':'partial_entry'})
```

```python
print all_rev
```

```python
Reviews=[]
for rev in all_rev:
    print rev.string
    Reviews.append(rev.string)
```

```python
dates= soup.find_all('span', attrs={'class','ratingDate relativeDate'})
```

```python
Date=[]
for date in dates:
    print date['title']
    Date.append(date['title'])
```

```python
names= soup.find_all('span', attrs={'class','expand_inline scrname'})
```

```python
Name=[]
for nam in names:
    print nam.string
    Name.append(nam.string)
```

```python
df=pd.DataFrame(Name,columns=['Name'])
```

```python
df["Date"]=Date
```

```python
df["Reviews"]=Reviews
```
